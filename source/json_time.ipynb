```python
stock_df.drop(columns=['Unnamed: 0', 'ticker','company_url','revenue','employee_count',
                       'sector','open_val','high_val','low_val', 'close_val',
                       'volume','volume_weight','number_of_transactions',
                        ], axis=1, inplace=True)
stock_df = stock_df.drop_duplicates(subset="company_name")
stock_df.reset_index(inplace=True)
stock_df.drop(columns="index", axis=1, inplace=True)
stock_df.head()
```

```python
stock_df.set_index("company_name").head()
```

```python
stock_df['coordinates'] = stock_df['latitude'].astype('str') + ", " + stock_df['longitude'].astype('str')
stock_df.drop(['latitude', 'longitude'], axis=1, inplace=True)
```

```python
stock_df.head()
```

```python
# stock_df['json'] = stock_df.apply(lambda x: x.to_json(), axis=1)
# stock_df['json']

geo_dict = stock_df.to_dict(orient='records')
# geo_json = json.load(d, ensure_ascii=False)
geo_dict[0]
```

```python
geo_json = json.dumps(geo_dict, indent=3)
print(geo_json)
```

```python
with open("../resources/company_all_star.json", "w") as outfile:
    outfile.write(geo_json)
```
