```python
# def df_to_geojson(stock_df, properties, lat='latitude', long='longitude', pc='percent_change', location='city_name' + ', ' + 'state_name'):
def df_to_geojson(stock_df, properties, lat='latitude', long='longitude', pc='percent_change', location='city_name', state='state_name'):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in stock_df.iterrows():
        feature = {'type':'Feature',
                   'properties':{'mag':[row[pc]], 'place':[row[location]]},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = [row[long],row[lat]]
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson
```

```python
cols = ['date_val', 'company_name', 'city_name', 'state_name', 'sector', 'region', 'country_code', 'percent_change', 'longitude', 'latitude']
geojson = df_to_geojson(stock_df, cols)
```
