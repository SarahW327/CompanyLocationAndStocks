```python
#import dependencies
import requests
import urllib.parse
import geopy
import pandas as pd
```

```python
stocks_df = pd.read_csv("./resources/stocks_data.csv")
stocks_df.head()
```

```python
stocks_df.columns
```

```python
stock_df = stocks_df.rename(columns={stocks_df.columns[0]: 'ticker', stocks_df.columns[1]: 'name', 
                                     stocks_df.columns[6]: 'city', stocks_df.columns[7]: 'state',
                                     stocks_df.columns[8]: 'country'},inplace=True)
stocks_df.head()
```

```python
stocks_df = stocks_df.drop(["LONGITUDE", "LATITUDE"], axis=1) 
stocks_df.head()
```

```python
cols = ["city", "state"]
city_state_s = stocks_df[cols].apply(lambda row: ", ".join(row.values.astype(str)), axis=1)
GeoSearch = pd.Series(city_state_s)
GeoSearch.head(15)
```

```python
# For loop with index should any coordinates cause error and easily identifiable 
# test code to make sure it works
long_lat = []
for idx, geo in enumerate(GeoSearch):
    address = geo
    url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(address) +'?format=json'

    response = requests.get(url).json()
    results = idx, response[0]["lat"], response[0]["lon"]
    long_lat.append(results)
    coordinates_s = pd.Series(long_lat)
    print(coordinates_s)
    #print(response[0]["lon"])
```

```python
coordinates_s.head(20)
```

```python
# separate longitude and latitude into different series and without index
longitude = []
latitude = []
for geo in GeoSearch:
    address = geo
    url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(address) +'?format=json'

    response = requests.get(url).json()
    results1 = response[0]["lon"]
    results2 = response[0]["lat"]
    longitude.append(results1)
    latitude.append(results2)
    longitude_s = pd.Series(longitude)
    latitude_s = pd.Series(latitude) 
```
